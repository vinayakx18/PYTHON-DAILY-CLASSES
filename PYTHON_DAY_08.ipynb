```python
# compile time / method overloading polymorphism

class BCA_1st_Semester :
  def student(self,a,b=0,c=0,d=0):
      return a+b+c+d

BCA = BCA_1st_Semester()
print(BCA.student(5))
print(BCA.student(10,7,8))
print(BCA.student(-10,4,9,-5))
print(BCA.student(99,4,-12))
```

```python
# Run time / Metthod overiding

class Animal :
      def sound(self):
        print("Animal makes a sound")

class Dog (Animal) :
      def sound(self):
        print("bow bow bow bow ")

class Cat(Animal) :
      def sound(self):
        print("meow meow meow meow")

class snake (Animal) :
      def sound(self) :
        print("buusssss")

class cow (Animal):
  def sound(self) :
    print("hanba hamba hamba")

obj1 = Animal()
obj2 = Dog()
obj3 = Cat()
obj4 = snake()
obj5 = cow()

obj1.sound()
obj2.sound()
obj3.sound()
obj4.sound()
obj5.sound()
```

```python
# operator overloading polymorphism

class book :
  def __init__(self,pages):
    self.pages = pages

  def __add__(self,other):
    return self.pages + other.pages

obj = book(200)
obj2 = book(300)
print("total pages is ",obj+obj2)
```

```python
# Abstraction example

from abc import ABC,abstractmethod
class vehicle(ABC) :
    def sound(self) :
      pass

class car (vehicle):
  def sound (self) :
    print ("I'm driving a car")

obj = car()
obj.sound()
```

```python
# Absrtaction Example

from abc import ABC,abstractmethod
class game(ABC) :


  # Normal function / method
    def free_fire(self) :
         print ("winner winner chicken dinner ")

  # Abstract function /method
    def BGMI (self):
      pass

  # Normal function / method
    def Ludo_king(self) :
         print ("I'm win the game")

class Temple_run (game):

  #BGMI function implemented by sub class
     def BGMI(self):
          print("I got Royal pass Diamond")

   # Normal function
     def Chess(self):
         print("I'm a district player")

obj = Temple_run()
obj.BGMI()
obj.Chess()
obj.free_fire()
```
